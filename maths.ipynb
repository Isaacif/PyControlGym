```python
import sympy as sp
import state_space as ss
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
from sympy import symbols, Eq, Function

init_vprinting()
t = sp.Symbol('t')
x1, x2 = dynamicsymbols('x_{1}(t) x_{2}(t)')
x1.diff()
```

```python
def x1_dot():
    eq = Eq(x1.diff(), x2)
    return eq

def x2_dot():
    m=0.1
    L=1
    b=0.05
    g=9.81
    x2_dot = -b/m*x2 -g/L*sp.sin(x1)
    eq = Eq(x2.diff(), -b/m*x2 -g/L*x1)
    return eq
```

```python
x1_dot(), x2_dot()
```

```python
x1_d = x1_dot()
x2_d = x2_dot()
Solver = ss.state_space("euler", *[x1_d, x2_d], symbolic=True)
Solver.runSimulation(0, 10, 0.01, *[0, 3], solve_method='ana')
Solver.plot2DResults('ana', 0)
```
